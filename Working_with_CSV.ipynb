# 7.Encoding Parameter

```python
#Let’s say your CSV file contains text like this:

Name,Country
José,Brasil
Renée,France


If the file was saved using UTF-8 encoding (the most common one), the “é” and “ó” are stored as special byte patterns that UTF-8 knows how to read.

But if your computer tries to read it as ASCII, it doesn’t know what those bytes mean — and you’ll see something like:

Jos�,Brasil
Ren�e,France


That’s a decoding error.
```

```python
So What Does the encoding Parameter Do?

In Pandas (and in Python generally), the encoding parameter tells your program which character encoding to use when reading a file.

Example:
import pandas as pd

df = pd.read_csv("data.csv", encoding="utf-8")


This means:

“Hey Pandas, interpret this file’s bytes using the UTF-8 character map.”
```

```python
When Should You Specify It?

✅ You usually don’t need to — Pandas assumes utf-8 by default.
But if you get an error like:

UnicodeDecodeError: 'utf-8' codec can't decode byte 0xe9 in position 10


That means Pandas guessed wrong — the file uses another encoding.

Try this:

df = pd.read_csv("data.csv", encoding="latin1")


…and it usually fixes it.
```

```python
How to Detect Encoding Automatically

If you’re not sure what encoding a file uses, you can detect it using the chardet library:

!pip install chardet
import chardet

with open("data.csv", "rb") as f:
    result = chardet.detect(f.read(10000))

print(result)


Output example:

{'encoding': 'ISO-8859-1', 'confidence': 0.73}


Then use:

df = pd.read_csv("data.csv", encoding="ISO-8859-1")
```
